## <b>Evolução de Empregos formais</b>

```python
# bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importando e lendo meus arquivos

jobs = pd.read_excel("Jobs.xlsx")
display(jobs)

faixa = pd.read_excel("Faixaetaria2.xlsx")
display(faixa)

ocupacao = pd.read_excel("ocupação.xlsx")
display(ocupacao)

taxa = pd.read_excel("taxarot.xlsx")
display(taxa)
```

```python
#Grafico do total de atividades por setor
plt.subplots(figsize=(20,10))
ax = sns.barplot(hue="Sexo", x="IBGE Setor", y="Atividade", data=jobs)

#valor flutuante
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height+.3,
            f'{round(p.get_height())}',
            ha="center",
            fontsize=12)

ax.set_title("Total de Atividade por Setor", fontsize=30) # titulo
ax.set_ylabel("Atividades", fontsize=20) # rotulo do eixo Y
ax.set_xlabel("Setores IBGE", fontsize=20) # rotulo do eixo X
ax.tick_params(axis='y', labelsize=13) # marcadores eixo Y
ax.tick_params(axis='x', labelsize=13) #marcadores eixo X

plt.show()
```

```python
# Empregos formais por faixa etaria
plt.subplots(figsize=(20,10))
ax2 = sns.barplot(hue="Sexo", x="Faixa Etaria", y="Dados", data=faixa)

#nomeros flutuantes
for p in ax2.patches:
    height = p.get_height()
    ax2.text(p.get_x()+p.get_width()/2.,
            height+.3,
            f'{round(p.get_height())}',
            ha = 'center',
            fontsize = 12)

ax2.set_title("Faixas Etarias por Setor 2020", fontsize=30) #titulo
ax2.set_ylabel("Pessoas", fontsize=20) # rotulo eixo Y
ax2.set_xlabel("Faixa Etaria", fontsize=20) # rotulo eixo X
ax2.tick_params(axis="y", labelsize=13) # marcadores eixo Y
ax2.tick_params(axis="x", labelsize=13) # marcadores eixo X

plt.show()
```

```python
#ocupação por setor
plt.subplots(figsize=(20,10))
ax3 = sns.barplot(hue="Sexo", x="CBO 2002 Ocupação", y="Dados", data=ocupacao)

#nomeros flutuantes
for p in ax3.patches:
    height = p.get_height()
    ax3.text(p.get_x()+p.get_width()/2.,
            height+.3,
            f'{round(p.get_height())}',
            ha = 'center',
            fontsize = 12)

ax3.set_title("Ocupação por Setor 2020", fontsize=30) #titulo
ax3.set_ylabel("Ocupação", fontsize=20) # rotulo eixo Y
ax3.set_xlabel("Setores", fontsize=20,) # rotulo eixo X
ax3.tick_params(axis="y", labelsize=13) # marcadores eixo Y
ax3.tick_params(axis="x", labelsize=13, rotation=45,) # marcadores eixo X

plt.show()
```

```python
#taxa de rotatividade 

plt.subplots(figsize=(20,10))
ax4 = sns.barplot(hue = "Setores", x="nivel geografico", y="Dados", data=taxa)

#nomeros flutuantes
for p in ax4.patches:
    height = p.get_height()
    ax4.text(p.get_x()+p.get_width()/2.,
            height+.3,
            f'{round(p.get_height(),2)}',
            ha = 'center',
            va = 'top',
            fontsize = 12)

ax4.set_title("Taxa de Rotatividade Dez 2019", fontsize=30) #titulo
ax4.set_ylabel("Taxa de rotatividade (%)", fontsize=20) # rotulo eixo Y
ax4.set_xlabel("Setores", fontsize=20,) # rotulo eixo X
ax4.tick_params(axis="y", labelsize=13) # marcadores eixo Y
ax4.tick_params(axis="x", labelsize=13, rotation=0,) # marcadores eixo X

plt.show()
```
